```
# Importando as bibliotecas necessárias

import pandas as pd
```

```
# Lendo dados diretamente no site da Câmara

# url2019 = 'http://dadosabertos.camara.leg.br/arquivos/votacoesVotos/csv/votacoesVotos-2019.csv'
# votacoes2019 = pd.read_csv(url2019, sep=';')
url2020 = 'http://dadosabertos.camara.leg.br/arquivos/votacoesVotos/csv/votacoesVotos-2020.csv'
votacoes2020 = pd.read_csv(url2020, sep=';')
# url2021 = 'http://dadosabertos.camara.leg.br/arquivos/votacoesVotos/csv/votacoesVotos-2021.csv'
# votacoes2021 = pd.read_csv(url2021, sep=';')
```

```
# Concatenando os dados num único DF
votacoes = votacoes2020.copy()
```

```
# Verificando os nomes das colunas

votacoes.columns.ravel()
```

```
# Excluindo colunas desnecessárias e reiniciando os índices

df = votacoes[['idVotacao', 'voto', 'deputado_id','deputado_nome', 'deputado_siglaPartido','deputado_siglaUf']]
df = df.reset_index(drop=True)
df
```

```
# Lendo arquivo de UFs x Regiões Geográficas IBGE

url = 'https://github.com/julianovale/PO450/blob/main/UFxREG.xlsx?raw=true'
ufxreg = pd.read_excel(url)
ufxreg
```

```
# Inserindo informação de Regiões no DF de votações, excluindo colunas desnecessárias,
# ordenando o DF e reiniciando os índices

df = df.merge(ufxreg,left_on='deputado_siglaUf',right_on='SIGLA_UF')
df = df[['idVotacao', 'voto', 'deputado_id','deputado_nome', 'deputado_siglaPartido','deputado_siglaUf','REG']]
df.sort_values(by=['idVotacao','deputado_id'],inplace=True)
df = df.reset_index(drop=True)
df
```

```
# Verificando NaN

df.isna().sum()
```

```
# Verificando tipos possíveis de voto

df.groupby('voto').count()
```
