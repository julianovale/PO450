```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
import requests
from io import BytesIO
```

```python
url = 'https://raw.githubusercontent.com/julianovale/PO450/main/dados.png'
imagem = requests.get(url)
```

```python
X = np.array(Image.open(BytesIO(imagem.content)).convert('L'))
plt.figure(figsize=(15,15))
plt.imshow(X, cmap='gray');
```

```python
rankX = np.linalg.matrix_rank(X)
UX, sX, VTX = np.linalg.svd(X)
```

```python
plt.stem(sX)
plt.show()
```
