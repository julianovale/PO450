## Exercício 2. a-c (função 1)



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = 5
mi = 0.01
t_iter = 400

it_hist = []
x_hist = []
fx_hist = []

for i in range(t_iter):
  x = x - mi*(2*x**3-6*x**2-4*x-1+50*np.cos(x))
  fx = 0.5*x**4-2*x**3-2*x**2-x+1+50*np.sin(x)
  it_hist.append(i)
  x_hist.append(x)
  fx_hist.append(fx)
```

```python
fx
```

```python
plt.plot(it_hist, x_hist)
plt.title("Evolução de x", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$x$", fontsize = 18, fontweight = 'bold');
```

```python
plt.plot(it_hist, fx_hist)
plt.title("Evolução de $f(x)$", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$f(x)$", fontsize = 18, fontweight = 'bold');
```

## Exercício 2. d.

Utilizando $\mu = 0.01$, $\mu = 0.001$ e $\mu = 0.0001$ para ambas a funções percebi grandes diferenças na evolução do $x$ para a função 1, sendo que em $\mu = 0.01$ o $x$ foi até para o lado oposto, crescendo. Pouca diferença no valor obtido para a função objetivo, atingindo respectivamente, -74.64999139455563, -74.6499913945556 e -74.64047686718766.<br/>
Para a função 2 os valores da função objetivo encontrados foram bem diversos e o "descida do gradiente" foi bem pronunciada, quanto maior o $\mu$ melhor. Respectivamente: 2.000005546601256, 12.079664163514554 e 44.60079713726775.
Nota-se também que à medida em que o $\mu$ se tornou menor, a curva de evolução dos valores de $x1$, $x2$ e $f(x1, x2)$ foi se tornando cada vez mais reta.
