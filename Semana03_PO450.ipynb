# Instalando e importando as bibliotecas necessárias

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from mpl_toolkits.mplot3d import Axes3D  
```

# Plotando as funções

```python
def fun1(x):
  return 0.5*x**4-2*x**3-2*x**2-x+1+50*np.sin(x)

t1 = np.arange(-3, 5, 0.01)


plt.plot(t1, fun1(t1), 'k');
```

```python
def fun2(x1, x2):
    return x1**2 + 0.5*x2**2 + 0.5*x1*x2 + 2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x1 = x2 = np.arange(-10.0, 10.0, 0.01)
X, Y = np.meshgrid(x1, x2)
zs = np.array(fun2(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z, cmap = 'rainbow')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
x = 5
mi = 0.01
t_iter = 400

it_hist = []
x_hist = []
fx_hist = []

for i in range(t_iter):
  x = x - mi*(2*x**3-6*x**2-4*x-1+50*np.cos(x))
  fx = 0.5*x**4-2*x**3-2*x**2-x+1+50*np.sin(x)
  it_hist.append(i)
  x_hist.append(x)
  fx_hist.append(fx)
```

```python
plt.plot(it_hist, x_hist)
plt.title("Evolução de x", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$x$", fontsize = 18, fontweight = 'bold');
```

```python
plt.plot(it_hist, fx_hist)
plt.title("Evolução de $f(x)$", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$f(x)$", fontsize = 18, fontweight = 'bold');
```

```python
x1 = x2 = 5
mi = 0.01
t_iter = 400

it_hist = []
x1_hist = []
x2_hist = []
fx1x2_hist = []

for i in range(t_iter):
  x1 = x1 - mi*(x1+x2)
  x2 = x1 - mi*(x2+x1)
  fx1x2 = x1**2+0.5*x2**2+0.5*x1*x2+2
  it_hist.append(i)
  x1_hist.append(x1)
  x2_hist.append(x2)
  fx1x2_hist.append(fx1x2)
```

```python
plt.plot(it_hist, x1_hist)
plt.title("Evolução de x1", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$x1$", fontsize = 18, fontweight = 'bold');
```

```python
plt.plot(it_hist, x2_hist)
plt.title("Evolução de x2", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$x2$", fontsize = 18, fontweight = 'bold');
```

```python
plt.plot(it_hist, fx1x2_hist)
plt.title("Evolução de $f(x1,x2)$", fontsize = 20, fontweight = 'bold')
plt.xlabel("Iterações", fontsize = 18, fontweight = 'bold')
plt.ylabel("$f(x1,x2)$", fontsize = 18, fontweight = 'bold');
```
