```
import pandas as pd
```

```
url2019 = 'http://dadosabertos.camara.leg.br/arquivos/votacoesVotos/csv/votacoesVotos-2019.csv'
votacoes2019 = pd.read_csv(url2019, sep=';')
url2020 = 'http://dadosabertos.camara.leg.br/arquivos/votacoesVotos/csv/votacoesVotos-2020.csv'
votacoes2020 = pd.read_csv(url2020, sep=';')
url2021 = 'http://dadosabertos.camara.leg.br/arquivos/votacoesVotos/csv/votacoesVotos-2021.csv'
votacoes2021 = pd.read_csv(url2021, sep=';')
```

```
frames = [votacoes2019, votacoes2020, votacoes2021]
votacoes = pd.concat(frames)
votacoes
```

```
votacoes.columns.ravel()
```

```
df = votacoes[['idVotacao', 'voto', 'deputado_id','deputado_nome', 'deputado_siglaPartido','deputado_siglaUf']]
df
```
