```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
```

```python
n, m = 20,30
A = np.random.normal(size=(n,m))
rankA = np.linalg.matrix_rank(A)

UA, sA, VTA = np.linalg.svd(A)

plt.imshow(A)
plt.show()
```

```python
plt.stem(sA)
plt.show()
```

```python
k = 5
G = np.random.normal(size=(n,k))
H = np.random.normal(size=(m,k))

B = np.dot(G, H.T)
rankB = np.linalg.matrix_rank(B)

plt.imshow(B)
plt.show()
```

```python
UB, sB, VTB = np.linalg.svd(B)
plt.stem(sB)
plt.show()
```

```python
X = np.array(Image.open('download.jpeg').convert('L'))
plt.imshow(X);
```

```python
rankX = np.linalg.matrix_rank(X)
UX, sX, VTX = np.linalg.svd(X)
```

```python
rankX
```

```python
plt.stem(sX)
plt.show()
```

```python
p = 10
UX_p = UX[:,:p]
VTX_p = VTX[:p]
sX_p = sX[:p]
SX_p = np.diag(sX_p)
```

```python
Xp = np.dot(UX_p,np.dot(SX_p,VTX_p))
```

```python
plt.imshow(Xp);
```
