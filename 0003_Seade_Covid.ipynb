```
# importando bibliotecas

import pandas as pd
```

```
# lendo dados da seade

bases = ['01_populacao.csv', '02_mortalidadeinfantil.csv', '03_obitos.csv', '04_PIB.csv', 
         '05_valoradicionado.csv', '06_consumoenergiaeletrica.csv', '07_veiculos.csv', 
         '08_nascidosvivos.csv', '09_leitosinternacao.csv', '10_profissionaissaude.csv', 
         '11_empregosformais.csv', '12_rendaempregos.csv', '13_idhdimensoes.csv', 
         '14_rendapercapita.csv', '15_agualixoesgosto.csv']

url = 'https://raw.githubusercontent.com/julianovale/PO450/main/seade/'

for i in range(len(bases)):
  if i == 0:
    df = pd.read_csv(url+bases[i], sep = ';', encoding= 'Windows-1252')
  else:
    df = df.merge(pd.read_csv(url+bases[i], sep = ';', encoding= 'Windows-1252'),left_on='Cód. IBGE',right_on='Cód. IBGE')
```
